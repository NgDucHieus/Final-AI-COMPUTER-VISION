```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Confusion matrix data: Replace this with your actual confusion matrix
conf_matrix = np.array([[54, 1, 3],
                        [0, 30, 2],
                        [3, 3, 49]])

# Define labels for the confusion matrix
labels = ['Unripe', 'Midripe', 'Ripe']

# Create a figure
plt.figure(figsize=(8, 6))  # Adjust figure size if needed

# Plot the confusion matrix using seaborn
heatmap = sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=labels, yticklabels=labels,
                      cbar=True, annot_kws={"size": 16})  # Adjust the font size of numbers with annot_kws

# Adjust colorbar font size
colorbar = heatmap.collections[0].colorbar
colorbar.ax.tick_params(labelsize=15)  # Change 14 to the desired font size for the colorbar ticks

# Add labels, title, and adjust font sizes
plt.xlabel("Predicted Label", fontsize=18)
plt.ylabel("True Label", fontsize=18)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

# Save the figure as a high-resolution image
plt.savefig("confusion_matrix_with_adjusted_colorbar.png", dpi=300, bbox_inches='tight')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the updated chart
categories = ['Khoa hc (30%)', 'Lch s (10%)', 'Ti chnh (5%)', 
              'Trit hc (12%)', 'Tiu thuyt & Khoa hc vin tng (30%)', 'Vn hc Vit Nam (13%)']
percentages = [30, 10, 5, 12, 30, 13]  # Adjusted percentages
colors = ['blue', 'skyblue', 'yellow', 'orange', 'red', 'lightblue']

# Create the pie chart
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(
    percentages,
    labels=None,  # Exclude labels from the pie chart
    colors=colors,
    startangle=90,
    autopct='%1.1f%%',  # Add numbers on the segments
    wedgeprops={'edgecolor': 'black', 'linewidth': 1}  # Adding stroke
)

# Customize the number styling on the chart
for autotext in autotexts:
    autotext.set_color('black')
    autotext.set_fontsize(10)

# Add a legend
plt.legend(wedges, categories, loc="center left", bbox_to_anchor=(1, 0.5), title="Categories")

# Adjust layout to prevent the legend from being cut off
plt.tight_layout()

# Save the figure with proper bounding box
plt.savefig("pie_fixed.jpg", bbox_inches='tight')

# Show the chart
plt.show()
```
